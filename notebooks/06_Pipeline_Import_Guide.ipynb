## 1️⃣ Estrutura dos Pipelines (.zip)

Cada pipeline é distribuído como um arquivo .zip contendo:

### Arquivos Inclusos:

```
pl_m_poc_xml_emp.zip
├── pl_m_poc_xml_emp/
│   ├── manifest.json                    (Metadados do pipeline)
│   └── pipeline_wf_m_poc_xml_emp.json   (Definição do pipeline)

pl_m_poc_xml_hr.zip
├── pl_m_poc_xml_hr/
│   ├── manifest.json                    (Metadados do pipeline)
│   └── pipeline_wf_m_poc_xml_hr.json    (Definição do pipeline)
```

### Conteúdo do manifest.json:
- **name**: Nome único do pipeline
- **displayName**: Nome amigável para exibição
- **description**: Descrição funcional
- **version**: Versão do pipeline
- **author**: Autor/Equipe responsável
- **environment**: Plataforma alvo (Fabric)
- **activities**: Lista de atividades/notebooks orquestrados

## 3️⃣ Processo de Importação

### Passo 1: Extrair o arquivo .zip
```
Location: C:\Users\[user]\Informatica-Scenarios\pipelines\
File: pl_m_poc_xml_emp.zip ou pl_m_poc_xml_hr.zip

Extrair para: pasta temporária (ex: ~/Desktop/pipelines)
```

### Passo 2: Abrir o Microsoft Fabric
1. Navegue até seu workspace no Fabric
2. Clique em **"+ New"** → **"Data Pipeline"**
3. Dê um nome descritivo (ex: "pl_m_poc_xml_emp")

### Passo 3: Importar o JSON do Pipeline
1. No editor do pipeline, clique em **"Import from JSON"**
2. Selecione o arquivo: `pipeline_wf_m_poc_xml_emp.json`
3. Verifique se as atividades foram importadas corretamente

### Passo 4: Referenciar os Notebooks
1. Para cada atividade do tipo Notebook:
   - Clique em **"Notebook"** (atividade)
   - Selecione **"Browse"** → escolha o notebook correto
   - Exemplo: `03_Map_EMP_Source_to_Target` ou `05_Map_HR_Source_to_Target`

### Passo 5: Configurar Parâmetros (se necessário)
1. Verifique os parâmetros de caminho de entrada/saída
2. Confirme que os locais do Lakehouse estão corretos:
   - Entrada: `/lakehouse/default/Files/employees.xml` ou `/lakehouse/default/Files/hr.xml`
   - Saída: `/lakehouse/default/Files/[output_folder]/`

## 4️⃣ Configuração e Execução

### Configuração de Retry
```json
{
  "retries": 3,
  "retry_wait_in_seconds": 30,
  "timeout_in_minutes": 60
}
```

2. Monitore o progresso em **"Monitor"**

## 5️⃣ Monitoramento e Troubleshooting

### Verificar Status do Pipeline
1. Acesse **"Monitor"** → **"Pipeline runs"**
2. Procure pela execução mais recente
3. Verifique o status de cada atividade

### Problemas Comuns

| Problema | Causa Provável | Solução |
|----------|---------------|---------|
| ❌ Notebook não encontrado | Referência quebrada | Re-selecionar notebook correto |
| ❌ Erro de permissão | Acesso ao Lakehouse | Verificar RBAC no Lakehouse |
| ❌ Timeout | Processamento lento | Aumentar timeout para 90+ minutos |
| ❌ Arquivo XML não encontrado | Caminho incorreto | Validar: `/lakehouse/default/Files/employees.xml` |
| ⚠️ Validação falha | Dados corrompidos | Verificar integridade do XML |

### Verificar Logs
1. Clique na atividade com falha
2. Abra **"Output"** para ver mensagens detalhadas
3. Procure por erros como:
   - `FileNotFound`: Validar caminho do arquivo
   - `ParseError`: Validar formato XML
   - `ValidationError`: Verificar esquema de dados
